```python
import shutil
import os

folders = ["train", "test", "valid"]
src_base = "tennis-ball-detection-6/tenis-ball-detection-6"
dst_base = "tennis-ball-detection-6/tennis-ball-detection-6"

for folder in folders:
    src = os.path.join(src_base, folder)
    dst = os.path.join(dst_base, folder)

    if os.path.exists(src):
        os.makedirs(os.path.dirname(dst), exist_ok=True)
        shutil.move(src, dst)
        print(f"Moved {src} -> {dst}")
    else:
        print(f"Source folder does NOT exist: {src}")
```
